### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = x ** 2

# Create the plot
plt.plot(x, y, label='y = x²', color='b', linewidth=2)

# Customize the plot
plt.title('Square Function Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values (x²)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```
